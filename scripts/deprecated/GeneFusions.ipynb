```python
import pandas as pandas
import numpy as np
```

```python
data = pandas.read_csv('/Users/carternorton/Desktop/chimerseq_7C2991727027ECF9130820C65C07E87E_1707359671387.txt', sep='\t')
```

```python
#Number of unique values in BarcodeID column
#Remove any information after the first 12 characters in the Barcode ID column
data['Barcode ID'] = data['Barcode ID'].str[:12]
data['Barcode ID'].nunique()
```

```python
#Phenodata
data2 = pandas.read_csv('/Users/carternorton/Desktop/RCC/KIRC_Phenotype.csv')

fusions = (data['Barcode ID'].unique().tolist())
#How many of barcode id are in TCGA sample code?
overlap = data2['TCGA Sample Code'].isin(fusions)

print(overlap.value_counts())
```

```python
#Let's select for mutations that are including in our phenotype data
data = data[data['Barcode ID'].isin(data2['TCGA Sample Code'])]

#If rows have a duplicated value in Fusion_pair and Barcode ID, remove the duplicated row
before = len(data)
data = data.drop_duplicates(subset=['Fusion_pair', 'Barcode ID'])
print('Number of duplicate fusions removed:', before - len(data))
print('Number of unique fusions:', len(data))
```

```python
#Let's make a list of counts of each fusion pair
full_fusions = data['Fusion_pair'].value_counts()
full_fusions.to_csv('/Users/carternorton/Desktop/RCC/fusion_data/full_fusions.csv')
```

```python
#Let's create a new column for the starting gene of the fusion pair
data['Gene1'] = data['Fusion_pair'].str.split('-').str[0]
data['Gene2'] = data['Fusion_pair'].str.split('-').str[1]

gene1_counts = data["Gene1"].value_counts()
gene2_counts = data["Gene2"].value_counts()

#which genes are in both gene1 and gene2?
gene1 = set(gene1_counts.index)
gene2 = set(gene2_counts.index)
common_genes = gene1.intersection(gene2)
print('Common genes:', common_genes)

#Let's save our gene counts
gene1_counts.to_csv('/Users/carternorton/Desktop/RCC/fusion_data/start_gene_counts.csv')
gene2_counts.to_csv('/Users/carternorton/Desktop/RCC/fusion_data/end_gene_counts.csv')
```
